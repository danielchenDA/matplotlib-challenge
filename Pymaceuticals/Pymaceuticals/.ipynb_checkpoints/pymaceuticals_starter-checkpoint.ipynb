```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_to_load)
clinical_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_df = pd.merge(mouse_df, clinical_df, on = "Mouse ID")

# Display the data table for preview
combined_df.head(20)
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
groupby_Drug_Timepoint = combined_df.groupby(["Drug","Timepoint"])
DT_mean_plist = groupby_Drug_Timepoint.mean()

# Convert to DataFrame
DT_mean_df = pd.DataFrame(DT_mean_plist)

# Preview DataFrame
DT_mean_df.head(40)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
DT_SE_df = groupby_Drug_Timepoint.sem()
# Preview DataFrame
DT_SE_df.head(10)
```

```python
# Minor Data Munging to Re-Format the Data Frames
DT_SE_df = DT_SE_df.rename(columns = {"Metastatic Sites" : "SE Metastatic Sites"})
DT_mean_df = DT_mean_df.rename(columns = {"Metastatic Sites" : "Mean Metastatic Sites"})

DT_SE_df = DT_SE_df.rename(columns = {"Tumor Volume (mm3)" : "SE Tumor Volume (mm3)"})
DT_mean_df = DT_mean_df.rename(columns = {"Tumor Volume (mm3)" : "Mean Tumor Volume (mm3)"})

DT_SE_df = DT_SE_df.drop(columns = "Mouse ID")

DT_SE_mean_df = pd.merge(DT_mean_df, DT_SE_df, left_index = True, right_index = True)

# Preview that Reformatting worked
DT_SE_mean_df.head()

```

```python
# Generate the Plot (with Error Bars)
fig, ax = plt.subplots()
plt.grid()

## Create Capomulin-specific DF and generate plot
capomulin_SE_mean_df = DT_SE_mean_df.iloc[DT_SE_mean_df.index.get_level_values("Drug") == "Capomulin"] 
capomulin_SE_mean_df.index = capomulin_SE_mean_df.index.droplevel("Drug")

ax.errorbar(capomulin_SE_mean_df.index.tolist(), 
            capomulin_SE_mean_df["Mean Tumor Volume (mm3)"].tolist(), 
            capomulin_SE_mean_df["SE Tumor Volume (mm3)"].tolist(),
            marker="v",
            markersize=4,
            linestyle="dashed",
            linewidth=1.5,
            label="Capomulin",
            color="green")

## Create Infubinol-specific DF and generate plot
infubinol_SE_mean_df = DT_SE_mean_df.iloc[DT_SE_mean_df.index.get_level_values("Drug") == "Infubinol"] 
infubinol_SE_mean_df.index = infubinol_SE_mean_df.index.droplevel("Drug")

ax.errorbar(infubinol_SE_mean_df.index.tolist(), 
            infubinol_SE_mean_df["Mean Tumor Volume (mm3)"].tolist(), 
            infubinol_SE_mean_df["SE Tumor Volume (mm3)"].tolist(),
            marker="s",
            markersize=4,
            linestyle="dashed",
            linewidth=1.5,
            label="Infubinol",
            color="orange")

## Create Ketapril-specific DF and generate plot
ketapril_SE_mean_df = DT_SE_mean_df.iloc[DT_SE_mean_df.index.get_level_values("Drug") == "Ketapril"] 
ketapril_SE_mean_df.index = ketapril_SE_mean_df.index.droplevel("Drug")

ax.errorbar(ketapril_SE_mean_df.index.tolist(), 
            ketapril_SE_mean_df["Mean Tumor Volume (mm3)"].tolist(), 
            ketapril_SE_mean_df["SE Tumor Volume (mm3)"].tolist(),
            marker="D",
            markersize=4,
            linestyle="dashed",
            linewidth=1.5,
            label="Ketapril",
            color="red")

## Create Placebo-specific DF and generate plot
placebo_SE_mean_df = DT_SE_mean_df.iloc[DT_SE_mean_df.index.get_level_values("Drug") == "Placebo"] 
placebo_SE_mean_df.index = placebo_SE_mean_df.index.droplevel("Drug")

ax.errorbar(placebo_SE_mean_df.index.tolist(), 
            placebo_SE_mean_df["Mean Tumor Volume (mm3)"].tolist(), 
            placebo_SE_mean_df["SE Tumor Volume (mm3)"].tolist(),
            marker="o",
            markersize=4,
            linestyle="dashed",
            linewidth=1.5,
            label="Placebo",
            color="blue")

# Add title, legend, y-axis label, and x-axis label
plt.title("Tumor Response to Treatment")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Treatment Duration (Days)")

plt.ylim(30, 75)
plt.xlim(-2, 50)

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, loc="upper left")

# Save the Figure
plt.savefig("Tumor Response to Treatment.png")


```

```python
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Count grouped by data and store in dataframe
count_mice_s = groupby_Drug_Timepoint["Mouse ID"].count()
count_mice_df = pd.DataFrame(count_mice_s)
count_mice_df = count_mice_df.rename(columns = {"Mouse ID":"Mouse Count"})

# Preview DataFrame
count_mice_df.head()
```

```python
# Generate the Plot (with Error Bars)
#fig, ax = plt.subplots()

# Create dataframes for each drug
## Capomulin
capomulin_mice_df = count_mice_df.iloc[count_mice_df.index.get_level_values("Drug") == "Capomulin"] 
capomulin_mice_df.index = capomulin_mice_df.index.droplevel("Drug")

capomulin_mice_list = capomulin_mice_df["Mouse Count"].tolist()
capomulin_mice_percent_list = [(100*(mice / capomulin_mice_list[0])) for mice in capomulin_mice_list]

## Infubinol
infubinol_mice_df = count_mice_df.iloc[count_mice_df.index.get_level_values("Drug") == "Infubinol"] 
infubinol_mice_df.index = infubinol_mice_df.index.droplevel("Drug")

infubinol_mice_list = infubinol_mice_df["Mouse Count"].tolist()
infubinol_mice_percent_list = [(100*(mice / infubinol_mice_list[0])) for mice in infubinol_mice_list]

## Ketapril
ketapril_mice_df = count_mice_df.iloc[count_mice_df.index.get_level_values("Drug") == "Ketapril"] 
ketapril_mice_df.index = ketapril_mice_df.index.droplevel("Drug")

ketapril_mice_list = ketapril_mice_df["Mouse Count"].tolist()
ketapril_mice_percent_list = [(100*(mice / ketapril_mice_list[0])) for mice in ketapril_mice_list]

## Placebo
placebo_mice_df = count_mice_df.iloc[count_mice_df.index.get_level_values("Drug") == "Placebo"] 
placebo_mice_df.index = placebo_mice_df.index.droplevel("Drug")

placebo_mice_list = placebo_mice_df["Mouse Count"].tolist()
placebo_mice_percent_list = [(100*(mice / placebo_mice_list[0])) for mice in placebo_mice_list]
```

```python
plt.grid()

# Create Line Graph
capomulin_handle, = plt.plot(capomulin_mice_df.index.tolist(), 
                            capomulin_mice_percent_list, 
                            marker="v",
                            linestyle="dashed",
                            linewidth=0.5,
                            label="Capomulin",
                            color="green")

infubinol_handle, = plt.plot(infubinol_mice_df.index.tolist(), 
                            infubinol_mice_percent_list, 
                            marker="s",
                            linestyle="dashed",
                            linewidth=0.5,
                            label="Infubinol",
                            color="orange")

ketapril_handle, = plt.plot(ketapril_mice_df.index.tolist(), 
                            ketapril_mice_percent_list, 
                            marker="D",
                            linestyle="dashed",
                            linewidth=0.5,
                            label="Ketapril",
                            color="red")

placebo_handle, = plt.plot(placebo_mice_df.index.tolist(), 
                            placebo_mice_percent_list, 
                            marker="o",
                            linestyle="dashed",
                            linewidth=0.5,
                            label="Placebo",
                            color="blue")

# Add title, legend, y-axis label, and x-axis label
plt.title("Mice Survival During Treatment")
plt.ylabel("Survival Percentage (%)")
plt.xlabel("Treatment Duration (Days)")

plt.ylim(30, 105)
plt.xlim(-2, 50)

plt.legend(handles=[capomulin_handle,infubinol_handle,ketapril_handle,placebo_handle],loc="best")

# Save the Figure
plt.savefig("Mice Survival Rate During Treatment.png")
```

```python
# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
## Capomulin
capomulin_mean_list = capomulin_SE_mean_df["Mean Tumor Volume (mm3)"].tolist()
capomulin_tumor_percent_list = [(100*(tumor / capomulin_mean_list[0]))-100 for tumor in capomulin_mean_list]

## Infubinol
infubinol_mean_list = infubinol_SE_mean_df["Mean Tumor Volume (mm3)"].tolist()
infubinol_tumor_percent_list = [(100*(tumor / infubinol_mean_list[0]))-100 for tumor in infubinol_mean_list]

## Ketapril
ketapril_mean_list = ketapril_SE_mean_df["Mean Tumor Volume (mm3)"].tolist()
ketapril_tumor_percent_list = [(100*(tumor / ketapril_mean_list[0]))-100 for tumor in ketapril_mean_list]

## Placebo
placebo_mean_list = placebo_SE_mean_df["Mean Tumor Volume (mm3)"].tolist()
placebo_tumor_percent_list = [(100*(tumor / placebo_mean_list[0]))-100 for tumor in placebo_mean_list]

## Bar Heights
tumor_change = (int(capomulin_tumor_percent_list[9]),
                  int(infubinol_tumor_percent_list[9]),
                  int(ketapril_tumor_percent_list[9]),
                  int(placebo_tumor_percent_list[9]))

## Bar x-axis
xaxis = np.arange(0,len(tumor_change),1)

# Display the data to confirm
tumor_change


```

```python
plt.grid()

# Create Line Graph
plt.bar(xaxis,
        height= tumor_change,
        color = ["Green","Red","Red","Red"])

# Create Bar ticks
plt.xticks(xaxis, ["Capomulin","Infubinol","Ketapril","Placebo"])

# Set Lims
plt.ylim(-25,60)

# Add aesthetics
plt.title("Tumor Growth during Treatment")
plt.xlabel("Treatment Drug")
plt.ylabel("Tumor Growth Percentage (%)")


plt.show()

# Save the Figure
plt.savefig("Tumor Growth during Treatment.png")

# Show the Figure
fig.show()
```
